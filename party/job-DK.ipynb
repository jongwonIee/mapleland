```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

def check_distribution_and_summary(df):
    """
    Check the distribution of `level` and job-map combinations to understand why so many rows are being filtered out.
    """
    # Check the distribution of level values
    print("\nDistribution of level values:")
    print(df['level'].describe())

    # Check how many rows are above the 99.9th percentile of level values (top 0.001%)
    level_threshold = np.percentile(df['level'], 1)  # Top 0.001% level value
    print(f"\nLevel threshold (90 percentile): {level_threshold}")
    rare_levels = df[df['level'] >= level_threshold]
    print(f"Rows with level above 99.9 percentile: {len(rare_levels)}")
    
    # Check the frequency of job-map combinations
    print("\nFrequency of job-map combinations:")
    job_map_counts = df.groupby(['job', 'map']).size()
    print(f"Top 10 job-map combinations:\n{job_map_counts.sort_values(ascending=False).head(10)}")
    print(f"\nNumber of job-map combinations used by more than 0.00001% of data: {len(job_map_counts[job_map_counts >= 0.00001 * len(df)])}")
    
# Example usage:
check_distribution_and_summary(df)
```
